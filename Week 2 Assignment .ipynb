### Question 5 : Given a list my_list = [1, 2, 3, 4, 5], write the code to slice the list and obtain the sub-list [2, 3]

```python
my_list=[1,2,3,4,5]
## slice Operation 
print(my_list[2:3])
print("Another Way is ",my_list[-3:-2])

## if you want 2,3 index string then 
print(my_list[2:4])
print("Another Way is ",my_list[-3:-1])
```
